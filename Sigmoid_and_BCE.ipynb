```python
import numpy as np 
import matplotlib.pyplot as plt
import math
import pandas as pd
import numpy.random as random
```

```python
def NormalizeData(data):
  return (data - np.min(data)) / (np.max(data) - np.min(data))


  
```

```python
x = np.vstack((np.random.randint(1,10, size=(10,2)),
             np.random.randint(20,30, size=(10,2)) ))

# to normalize data between 1 and 0
x = NormalizeData(x)

y = np.hstack(( np.zeros(10, dtype=int), np.ones(10, dtype=int))).reshape(-1,1)
```

```python
def forward_pass(x, w1, b1, w2, b2, y):
  eps = 1e-6
  hidden_1 = x.dot(w1.T) + b1
  hidden_1_sig = 1/(1 + np.exp(-hidden_1))
  
  hidden_2 = hidden_1_sig.dot(w2) + b2
  hidden_2_sig = 1/(1 + np.exp(-hidden_2) + eps)

  Loss = -np.mean(np.multiply(y, np.log(hidden_2_sig)) + np.multiply( (1-y), np.log(1 - hidden_2_sig)))

  print(" the loss: ", Loss.sum())
  
  return hidden_1, hidden_1_sig, hidden_2, hidden_2_sig





```

```python
# weights initialization 
w1 = random.uniform(-1,1,(100,2))
b1 = random.uniform(-1,1,(1,100))
w2 = random.uniform(-1,1,(100,1))
b2 = random.uniform(-1,1,(1))
```

```python
def backward_pass(x, hidden_2_sig, hidden_2, hidden_1_sig, hidden_1, y):
  dLdw2 = (hidden_2_sig - y).T.dot(hidden_1_sig)
  dLdw1 = ( ( ( hidden_2_sig - y).dot(w2.T)) * (np.multiply(hidden_1_sig, (1-hidden_1_sig)))).T.dot(x)
  dLdb2 = (hidden_2_sig - y).sum(axis=0)
  dLdb1 = ((hidden_2_sig-y).dot(w2.T) * (np.multiply(hidden_1_sig, (1-hidden_1_sig)))).sum(axis=0)

  return dLdw2, dLdw1, dLdb2, dLdb1

```

```python
def update_parameters(w1, w2, dw2, dw1, b1, b2, db1, db2, learning_rate):

    N_w1 = w1 - dw1*learning_rate
    N_w2 = w2 - dw2*learning_rate
    N_b1 = b1 - db1*learning_rate
    N_b2 = b2 - db2*learning_rate

    return N_w1, N_w2, N_b1, N_b2



```

```python
for i in range(100):
  hidden_1, hidden_1_sig, hidden_2, hidden_2_sig = forward_pass(x, w1, b1, w2, b2, y)
  dw2, dw1, db2, db1 = backward_pass(x, hidden_2_sig, hidden_2, hidden_1_sig, hidden_1, y )
  learning_rate = 1e-3
  
  N_w1, N_w2, N_b1, N_b2 = update_parameters(w1,w2,dw2,dw1,b1,b2,db1,db2,learning_rate)
  w1, w2, b1, b2 = N_w1, N_w2, N_b1, N_b2





```
